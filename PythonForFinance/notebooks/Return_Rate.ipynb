```python
import pandas as pd
import numpy as np
import plotly.express as px
```

# Retorno de ações individuais

## ITUB


```python
## Retorno Simples

preco_final = dataset['ITUB4'][len(dataset) -1 ] #ultimo valor
preco_inicial = dataset['ITUB4'][0] #primeiro valor

taxa_retorno_S = ((preco_final - preco_inicial) / preco_inicial)*100 ## usando fórmula de taxa simples
print(taxa_retorno_S)
```

```python
# Taxa de retorno do dia anterior
# para fazer o calculo do dia anterior, utilizamos o shift para buscar o preço final (function shift pula a quantidade do valor passado)
dataset['RSD ITUB4'] = (dataset['ITUB4'] / dataset['ITUB4'].shift(1)) - 1
```

```python
# Taxa de Retorno Simples Anual
(dataset['RSD ITUB4'].mean() * 246 )* 100
```

```python
# Taxa de retorno Logarítimica 
np.log(dataset['ITUB4'][len(dataset) -1 ] /dataset['ITUB4'][0]) * 100
```

```python
# Taxa de RL diária 
dataset['RL ITUB4'] = np.log(dataset['ITUB4'] / dataset['ITUB4'].shift(1))
```

```python
# Visualizando o gráfico para descobrir baixas e altas das ações 
dataset['RL ITUB4'].plot()
```

```python
# Média 
dataset['RL ITUB4'].mean()
```

```python
# RL anual
(dataset['RL ITUB4'].mean() * 246) *100
```

# Retorno de uma Carteira

```python
dataset = pd.read_csv('acoes.csv')
if 'Unnamed: 0' in dataset.columns:
    dataset.drop(columns=['Unnamed: 0'], inplace=True)
dataset
```

```python
dataset_normalizado = dataset.copy()
for i in dataset_normalizado.columns[1:]:
    dataset_normalizado[i] = (dataset_normalizado[i] / dataset_normalizado[i][0])
dataset_normalizado
```

```python
dataset_normalizado.plot(x= 'Date', figsize= (18,8));
```

```python
dataset_normalizado.drop(labels='Date', axis=1, inplace= True)
```

```python
#retorno simples pois queremos comparar várias ações
retorno_carteira = (dataset_normalizado/dataset_normalizado.shift(1)) - 1 
```

```python
retorno_anual = (retorno_carteira.mean() * 256) * 100
retorno_anual
```

```python
#Definir pesos para ação (de 100 % das suas ações, quanto % cada ação vale dele?)
# Taxa de retorno = (RetornoSimples * pesoDaAção) 
# Taxa de retorno da carteirea é só somar cada uma das taxa de retorno
peso_carteira1 = np.array([0.1, 0.1, 0.2, 0.1, 0.2, 0.1, 0.1, 0.1, 0.0])
```

```python
np.dot(retorno_anual, peso_carteira1)
```

# Comparativo Retorno de Carteira X IBOVESPA

```python
dataset = pd.read_csv('acoes.csv')
if 'Unnamed: 0' in dataset.columns:
    dataset.drop(columns='Unnamed: 0',inplace=True )
```

```python
dataset_normalizado = dataset.copy()
for i in dataset_normalizado.columns[1:]:
    dataset_normalizado[i] /= dataset_normalizado[i][0] 
dataset_normalizado.head()
```

```python
#Criando a coluna Carteira

dataset_normalizado['CARTEIRA'] = (dataset_normalizado['BBSE3'] + dataset_normalizado['ITUB4'] + dataset_normalizado['BEEF3'] + dataset_normalizado['CMIG4'] + dataset_normalizado['ITSA4'] + dataset_normalizado['PETR3'] + dataset_normalizado['PETR4'] + dataset_normalizado['POSI3']) / 5 
```

```python
figura = px.line(title= 'Comparativo Carteira x BOVA')
for i in dataset_normalizado.columns[1:]:
    figura.add_scatter(x= dataset_normalizado['Date'], y= dataset_normalizado[i], name=i)
figura.show()
```
